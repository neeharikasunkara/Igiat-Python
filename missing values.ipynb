# Missing values:
    fillna
    dropna- used to delete the line

```python
import pandas as pd
df=pd.read_csv('emp.csv')
df
```

```python
df.fillna(0)
```

```python
df.fillna('noname')
```

```python
## Filling of values using dictionary
df.fillna(value={'ename':'no name','eaddress':'no address','esalary':0})
```

```python
df.dropna(thresh=1) ##thresh=1 means prints if atleast there is 1 NAN
```

```python
df.dropna()  ## removes all rows which are consisting of NAN
```
