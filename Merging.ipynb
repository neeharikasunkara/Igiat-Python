```python
import pandas as pd
```

```python
d1={
    "id":[1,2,3,4],
    "Names":["abc","def","ghi","jkl"],
    "Location":["Vizag","anakapalle","Vizianagaram","Srikakulam"]}
df1=pd.DataFrame(d1)
df1
```

```python
d2={
    "id":[5,6,7,8],
    "Names":["mno","pqr","stu","vwx"],
    "Location":["Anakapalle","Vizag","Vizianagaram","Vizag"],
    "Qualification":["B.Tech","MCA","Diploma","Btech"]
}
df2=pd.DataFrame(d2)
print(df2.to_string(index=False))
```

```python
pd.merge(df1,df2, how="outer")   ### Outer join // Outer merge
```

```python
pd.merge(df1,df2, how="inner",on="Location")      ### Inner Join(Joins common things)
```

```python
pd.merge(df1,df2, how="inner")
```

```python
pd.merge(df1,df2, how="left",on="Location")    
#### Only df1(left) data will be there and printed first and
### df2(right)data will be printed if there is any match only
```

```python
pd.merge(df1,df2, how="right",on="Location")  
###Right data is printed and left common data is printed
```
